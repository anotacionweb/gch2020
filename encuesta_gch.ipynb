```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel('Anotación web colaborativa(2020-08-04).xlsx')

data = data.replace(['Acuerdo', 'Acuerdo total'], 'Sí')
data = data.replace(['Indiferencia', 'Desacuerdo', 'Desacuerdo total'], 'No')
```

```python
df = data[[column for column in data.columns if column.startswith('Tener que anotar los artículos seleccionados me ayudó a...')]]
df.columns = [column.split('>> ')[-1] for column in df.columns]
df = df.apply(pd.Series.value_counts)
df.loc[['Sí', 'No']].T.plot(kind='barh', stacked=True)
```

```python
df = data[[column for column in data.columns if column.startswith('Las anotaciones de mis compañerxs me ayudaron a...')]]
df.columns = [column.split('>> ')[-1] for column in df.columns]
df = df.apply(pd.Series.value_counts)
df.loc[['Sí', 'No']].T.plot(kind='barh', stacked=True)
```

```python
df = data[[column for column in data.columns if column.startswith('En general, las anotaciones...')]]
df.columns = [column.split('>> ')[-1] for column in df.columns]
df = df.apply(pd.Series.value_counts)
df.loc[['Sí', 'No']].T.plot(kind='barh', stacked=True)
```

```python
df = data[[column for column in data.columns if column.startswith('En este contexto de clases virtuales, siento que la actividad de anotación web colaborativ...')]]
df.columns = [column.split('>> ')[-1] for column in df.columns]
df = df.apply(pd.Series.value_counts)
df.loc[['Sí', 'No']].T.plot(kind='barh', stacked=True)
```

```python
selection = [
    '¿Te gustaría volver a usar la anotación web en algún próximo curso que tomes?',
    '¿Usarías la anotación web profesional o personalmente para hacer anotaciones privadas?',
    '¿Usarías la anotación web profesional o personalmente para anotar textos colaborativamente?'
]
df = data[selection]
df.columns = [column.split('>> ')[-1] for column in df.columns]
df = df.apply(pd.Series.value_counts)
df.loc[['Sí', 'No']].T.plot(kind='barh', stacked=True)
```

```python
df = data[[column for column in data.columns if column.startswith('¿Encontraste útiles los recursos compartidos?')]]
df.columns = [column.split('>> ')[-1].split(' (<a')[0][:50] for column in df.columns]
df = df.replace(['Útil', 'Muy útil'], 'Sí')
df = df.replace(['Poco útil'], 'No')
df = df.replace(['No lo consulté'], 'Sin consultar')
df = df.apply(pd.Series.value_counts)
df.loc[['Sí', 'No', 'Sin consultar']].T.plot(kind='barh', stacked=True)
```

```python
df = data[[column for column in data.columns if column.startswith('En la actividad 2, cuán fácil/difícil te resultó:')]]
df.columns = [column.split('>> ')[-1] for column in df.columns]
df = df.apply(pd.Series.value_counts)
df.loc[['Muy fácil', 'Fácil']].T[::-1].plot(kind='barh', stacked=True)
```
